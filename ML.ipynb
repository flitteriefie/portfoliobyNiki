```python
from sklearn.preprocessing import LabelEncoder
labe = LabelEncoder()
df['Age_Cat'] = labe.fit_transform(df['Age'])
df['Sex_Cat'] = labe.fit_transform(df['Sex'])
df['Race_Cat'] = labe.fit_transform(df['Race'])
df['DeathCity_Cat'] = labe.fit_transform(df['DeathCity'])
df['Heroin_Cat'] = labe.fit_transform(df['Heroin'])
df['Cocaine_Cat'] = labe.fit_transform(df['Cocaine'])
df['Fentanyl_Cat'] = labe.fit_transform(df['Fentanyl'])
df['FentanylAnalogue_Cat'] = labe.fit_transform(df['FentanylAnalogue'])
df['Oxycodone_Cat'] = labe.fit_transform(df['Oxycodone'])
df['Oxymorphone_Cat'] = labe.fit_transform(df['Oxymorphone'])
df['Ethanol_Cat'] = labe.fit_transform(df['Ethanol'])
df['Hydrocodone_Cat'] = labe.fit_transform(df['Hydrocodone'])
df['Benzodiazepine_Cat'] = labe.fit_transform(df['Benzodiazepine'])
df['Methadone_Cat'] = labe.fit_transform(df['Methadone'])
df['Amphet_Cat'] = labe.fit_transform(df['Amphet'])
df['Tramad_Cat'] = labe.fit_transform(df['Tramad'])
df['Morphine_NotHeroin_Cat'] = labe.fit_transform(df['Morphine_NotHeroin'])
df['Hydromorphone_Cat'] = labe.fit_transform(df['Hydromorphone'])
df['Other_Cat'] = labe.fit_transform(df['Other'])
df['OpiateNOS_Cat'] = labe.fit_transform(df['OpiateNOS'])
df['AnyOpioid_Cat'] = labe.fit_transform(df['AnyOpioid'])
df['MannerofDeath_Cat'] = labe.fit_transform(df['MannerofDeath'])
df['NamedDrug_Cat'] = labe.fit_transform(df['NamedDrug'])
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Drug Overdose Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
from sklearn.preprocessing import MinMaxScaler
x = data.values #returns a numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx [0:5]
```

```python
dist_matrix = distance_matrix(feature_mtx,feature_mtx) 
print(dist_matrix)
```

```python
agglom = AgglomerativeClustering(n_clusters = 7, linkage = 'complete')
agglom.fit(feature_mtx)
agglom.labels_
```

```python
data['cluster_'] = agglom.labels_
data.head()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=data['cluster_'], cmap='rainbow')
```

```python
data.groupby(['cluster_'])['cluster_'].count()
```

```python
agg_data = data.groupby(['cluster_'])['Age_Cat', 'Sex_Cat','Race_Cat', 'DeathCity_Cat', 'NamedDrug_Cat',
                                     'Morphine_NotHeroin_Cat', 'Other_Cat','OpiateNOS_Cat', 'AnyOpioid_Cat', 'MannerofDeath_Cat'].mean()
agg_data
```

```python
import matplotlib.cm as cm
n_clusters = max(agglom.labels_)+1
colors = cm.rainbow(np.linspace(0, 1, n_clusters))
cluster_labels = list(range(0, n_clusters))

# Create a figure of size 6 inches by 4 inches.
plt.figure(figsize=(50,50))

#plt.figure(figsize=(50,50))
for color, label in zip(colors, cluster_labels):
    subset = agg_data.loc[(label,),]
    for i in subset.index:
        plt.text(subset.loc[i][0]+5, subset.loc[i][2], 'Age_Cat='+str(int(i)) + ', Sex_Cat='+str(int(subset.loc[i][3]))+'k')
    plt.scatter(subset.Race_Cat, subset.DeathCity_Cat, s=subset.Age_Cat*20, c=color, label='cluster'+str(label))
plt.legend()
plt.title('Clusters')
plt.xlabel('??')
plt.ylabel('?')
```
